# First Up, Logistic Regression

We begin by creating X (independent) and y(dependent) Arrays. 

```python
print("First we'll create dummy variables for two categorical variables Family Size and Education")

#pd.concat([CampData_v2, pd.get_dummies(CampData_v2['Family'])], axis=1)
CampData_v3=pd.concat([CampData_v2, pd.get_dummies(CampData_v2['Education'])], axis=1)
CampData_v3=CampData_v3.rename(columns={1:'undergrad',2:'postgrad',3:'advanced'})
CampData_v3=pd.concat([CampData_v3, pd.get_dummies(CampData_v2['Family'])], axis=1)
CampData_v3=CampData_v3.rename(columns={1:'family_1',2:'family_2',3:'family_3',4:'family_4'})
print("\n Also let's drop the independent variables which were found to have no relationship with the dependent")
CampData_v3=CampData_v3.drop(['Education','Family','Online','CreditCard','Securities Account','Experience','Age'], axis=1)
```

```python

print ("Firstly we'll develop the model on the original data without the dummies that we created")
X = CampData_v2.iloc[:,1:12]
y = CampData_v2.iloc[:,0]
print("Now we'll divide the data into test and train in a 70:30 split")
```

```python
test_size = 0.30 # taking 70:30 training and test set
seed =1 # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
print("Now let's fit the model")
```

```python
model = LogisticRegression()
model.fit(X_train, y_train)
y_predict = model.predict(X_test)
model_score = model.score(X_test, y_test)
print("The model accuracy is: " + str(model_score))

print ("\n The other model stats are as follows \n")
print(metrics.classification_report(y_test, y_predict))
```

```python
print("The confusion matrix is as follows")
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(metrics.confusion_matrix(y_test, y_predict)), annot=True, cmap="PuOr" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```

#### The model accuracy is high at 94.9%. However, if we look at the precision and recall of the personal loan buyers, there's significant room for improvement. This is also depiced by the F1 score.
***

> - __True Positives__ : On the test data we correctly predicted Personal Loan takers = 88
> - __True Negative__ : On the test data we correctly predicted non-Personal Loan takers = 1336
> - __False Positives__ : On the test data we incorrectly predicted non-Personal Loan Takers as Personal Loan takers = 15
> - __False Negatives__ : On the test data we incorrectly predicted Personal Loan Takers as non-Personal Loan takers = 61
> - __Accuracy__ : (1336 + 88)/1500 = __94.9%__
> - __Recall__ : 88/(88 + 61) = __59%__
> - __Precision__ = 88/(88+15) = __85.4%__
> - __False Positive Rate__ = 15/(15+1336) = __1.1%__
> - __F1 Score__ = (2 x (88/103) x (88/149))/((88/103) + (88/149)) = __70%__
***

#### Okay, now let's try this with the dummy variables and only important variables that we created


```python
X = CampData_v3.iloc[:,1:13]
y = CampData_v3.iloc[:,0]
print('Let us also scale this data')
test_size = 0.30 # taking 70:30 training and test set
seed =1 # Random numbmer seeding for reapeatability of the code
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=seed)
X_train_scaled = preprocessing.scale(X_train)
X_test_scaled = preprocessing.scale(X_test)
model = LogisticRegression()
model.fit(X_train_scaled, y_train)
y_predict = model.predict(X_test_scaled)
model_score = model.score(X_test_scaled, y_test)
print("\n The model accuracy is: " + str(model_score))
print(metrics.classification_report(y_test, y_predict))
```

```python
print("The confusion matrix is as follows")
class_names=[0,1] # name  of classes
fig, ax = plt.subplots()
tick_marks = np.arange(len(class_names))
plt.xticks(tick_marks, class_names)
plt.yticks(tick_marks, class_names)
# create heatmap
sns.heatmap(pd.DataFrame(metrics.confusion_matrix(y_test, y_predict)), annot=True, cmap="Blues" ,fmt='g')
ax.xaxis.set_label_position("top")
plt.tight_layout()
plt.title('Confusion matrix', y=1.1)
plt.ylabel('Actual label')
plt.xlabel('Predicted label')
plt.show()
```
